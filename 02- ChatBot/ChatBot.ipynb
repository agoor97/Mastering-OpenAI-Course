```python
## Import Libraries, and read the key-token
import openai
import os
from dotenv import load_dotenv
import time

## Load the Key token
_ = load_dotenv()
key_token = os.getenv('OpenAI_KEY_TOKEN')

## Assign that key_token to api_key of OpenAI
openai.api_key = key_token
```

```python
## BaseLine Model, Trying makeing a chatbot

## User prompt Questions
uesr_q1 = 'What are the healthiest foods to eat every day?'
user_q2 = 'Make a recipe with those foods'
user_q3 = 'What is the amount of kcal?'


## Define System prompt 
system_prompt = '''Answer as concisely as possible.'''


## For Question-1 (user_q1)
messages = [ 
    {'role': 'system', 'content': system_prompt},    ## The system prompt
    {'role': 'user', 'content': uesr_q1}             ## The first question
        ]
response = openai.ChatCompletion.create(
              model='gpt-3.5-turbo',           
              messages=messages,
              temperature=0.7,  
              max_tokens=100,   
                )
response_q1 = response['choices'][0]['message']['content']
print(response_q1)
print('***'*40)
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++ ##

## For Question-2 (user_q2)
messages = [ 
    {'role': 'system', 'content': system_prompt},   ## The system prompt
    {'role': 'user', 'content': uesr_q1},           ## The first question
    {'role': 'assistant', 'content': response_q1},  ## the response of first question
    {'role': 'user', 'content': user_q2}            ## the second question
        ]
response = openai.ChatCompletion.create(
              model='gpt-3.5-turbo',           
              messages=messages,
              temperature=0.7,   
              max_tokens=100,   
                )
response_q2 = response['choices'][0]['message']['content']
print(response_q2)
print('***'*40)
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++ ##

## For Question-3 (user_q3)
messages = [ 
    {'role': 'system', 'content': system_prompt},   ## The system prompt
    {'role': 'user', 'content': uesr_q1},           ## The first question
    {'role': 'assistant', 'content': response_q1},  ## the response of first question
    {'role': 'user', 'content': user_q2},           ## the second question
    {'role': 'assistant', 'content': response_q2},  ## the response of second question
    {'role': 'user', 'content': user_q3}            ## The third question
        ]  
response = openai.ChatCompletion.create(
              model='gpt-3.5-turbo',           
              messages=messages,
              temperature=0, 
              max_tokens=100,   
                )
response_q3 = response['choices'][0]['message']['content']
print(response_q3)
```

### `Looping`

```python
## Looping to make it much more easier
all_prompts = list()
all_messages= list()
model_responses = list()

## Create system prompt
system_prompt = 'Answer as concisely as possible.'
all_messages.append({'role': 'system', 'content': system_prompt})

## Looping while true
while True:
    user_prompt = input('You: ')
    
    ## If the user wants to exit the chatbot -> break
    if user_prompt.lower() in ['quit', 'exit', 'ex', 'out', 'escape']:
        print('Thanks for using my ChatBot')
        time.sleep(2)  ## wait 2 seconds
        break
    ## If the user doesn't write any thing -> Continue
    elif user_prompt.lower() == '': 
        continue

    ## Answer
    else:
        ## append the question of user to message as a user roke
        all_messages.append({'role': 'user', 'content': user_prompt})
        
        ## also the question to questions list
        all_prompts.append(user_prompt)

        ## Call the API
        each_response = openai.ChatCompletion.create(
                model='gpt-3.5-turbo',           
                messages=all_messages,
                temperature=0.7,  
                max_tokens=1000,   
                    )
        each_response = each_response['choices'][0]['message']['content']
        print('ChatBot: ', each_response)  ## print the response of the api

        ## also add the resposne to model_responses list
        model_responses.append(each_response)  

        ## We must append this respond to the messages
        all_messages.append({'role': 'assistant', 'content': each_response})
        
        ## some space
        print('\n', '****'*50)
```
